```python
import pandas as pd
df = pd.read_csv("datasets/presidents.csv")
df.head(10)
```

```python
df['First'] = df['President']
df.head()
```

```python
df['First'] = df['First'].replace("[ ].*",'',regex=True)
df.head()
```

```python
df["President"] = df["President"].replace(".*[ ]",'',regex=True)
df.columns
df = df.rename(columns = {'President':'Last'})
df.head()
```

```python
df["Age at Term Start"] = df['Age atstart of presidency']
df["Age at Term Start"] = df["Age at Term Start"].replace("[A-Z].*","",regex=True)
df["Age atstart of presidency"] = df['Age atstart of presidency'].replace(r"^.*?(?=[A-Z])","",regex=True)
df['Born'] = df["Born"].replace("\[a\]?","",regex=True)
df = df.rename(columns = {"Age atstart of presidency":"Term Start Date"})
df.head()
```

```python
df = df.rename(columns= {"Post-presidencytimespan":"Lifespan post presidency","Age":"Age at death"})
df.head()
```

```python
print(df.columns)
```

```python
df["Age at Term End"] = df["Age atend of presidency"]
df["Age at Term End"] = df["Age at Term End"].replace("[A-Z].*","",regex=True)
df["Age atend of presidency"] = df["Age atend of presidency"].replace(r"^.*?(?=[A-Z])","",regex=True)
df = df.rename(columns = {"First":"First Name","Last":"Last Name","Age atend of presidency":"Term End Date"})
df.head()
```

```python
print(list(df.columns))
```

```python
df_reordered = df[['#', 'First Name','Last Name', 'Born', 'Term Start Date', 'Age at Term Start', 'Term End Date', 'Age at Term End', 'Lifespan post presidency', 'Died', 'Age at death']]
df_reordered
```

```python
df["Term End Date"] = df["Term End Date"].replace("\[.\]",'',regex=True)
df["Lifespan post presidency"] = df["Lifespan post presidency"].fillna("0 days")
df = df[['#', 'First Name','Last Name', 'Born', 'Term Start Date', 'Age at Term Start', 'Term End Date', 'Age at Term End', 'Lifespan post presidency', 'Died', 'Age at death']]
df[["Term Start Date","Term End Date","Died"]] = df[["Term Start Date","Term End Date","Died"]].apply(pd.to_datetime, errors = 'coerce')
df["Died"] = df["Died"].fillna("Still alive")
df
```
